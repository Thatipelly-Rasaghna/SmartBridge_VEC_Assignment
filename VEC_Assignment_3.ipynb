```python
#converting categorical labels in to numerical labels is called label encoding
from sklearn.preprocessing import LabelEncoder
```

```python
#initializing the label encoding
le=LabelEncoder()
```

```python
#converting a categorical column into numerical by using fit and transform
df['Species']=le.fit_transform(df['Species'])
```

```python
#droping the unwanted columns
df1=df.drop(columns=['Id'],axis=1)
```

```python
#seperating th data
#dependent and independent variables
#x is an independent variable
x=df1.drop(['Species'],axis=1)
```

```python
#y is an dependent variable
y=df1['Species']
```

```python
x
```

```python
y
```

```python
#spliting the data for training and testing
#importing the train and test split from sklearn
from sklearn.model_selection import train_test_split
```

```python
#initializing the variables x-train,y-train
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
#importing the model
# model selection has to done
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
```

```python
#initialising the model
lr=LogisticRegression()
```

```python
#fit or train the model
lr.fit(x_train,y_train)
```

```python
#to predict the model
pred=lr.predict(x_test)
```

```python
#to find the accuracy of the given model
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,pred)
```

```python
#importing a model
#SUPPORT VECTOR MACHINE(SVM) IS USED
#SVM LINEAR
from sklearn import svm
```

```python
#initializing a model
model=svm.SVC(kernel="linear")
```

```python
# fit/training the model
Md=model.fit(x_train,y_train)
```

```python
#predicting the model
pred=Md.predict(x_test)
```

```python
#import the metrics to find accurcy of a model
from sklearn import metrics
```

```python
metrics.accuracy_score(pred,y_test)
```
